```python
# import libraries
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# name of desired ticker
ticker = 'SPY'

# get the ticker of your desired ETF (use "^GSPC" if you want the S&P 500 index itself)
stock_data = yf.Ticker(ticker)

# get the historical market data for the provided ticker; adjust timeframe as desired
d_hist = stock_data.history(period = '10y')
```

```python
# display the first 5 rows of the dataframe (in ascending order from beginning of time period)
d_hist.head()
```

```python
# emphasize the ticker's closing price at the beginning of the time period
d_hist['Close'].iloc[0]
```

```python
# display the last 5 rows of the dataframe
d_hist.tail()
```

```python
# emphasize and store the ticker's closing price today
d_hist['Close'].iloc[len(d_hist.index) - 1]
```

```python
# plot the graph over the short term (to compare to the long term)
short_d_hist = stock_data.history(period = '1y')
sns.set(rc = {'figure.figsize':(15,8)}, font_scale = 1.5)
sns.lineplot(data=short_d_hist, x='Date', y='Close')
plt.title(f'Closing Prices for {ticker} Over the Short Term')
plt.xlabel('Year')
plt.ylabel('Closing Price (USD)')
plt.show()
```

```python
# plot the graph over the long term
sns.set(rc = {'figure.figsize':(15,8)}, font_scale = 1.5)
sns.lineplot(data=d_hist, x='Date', y='Close')
plt.title(f'Closing Prices for {ticker} Over the Long Term')
plt.xlabel('Year')
plt.ylabel('Closing Price (USD)')
plt.show()
```

# Conclusion

While broad market ETFs may exhibit significant volatility and decline in the short term, over the long term they have always gone up in accordance with the indices they track. As supported by history, bull markets are more frequent and last longer than bear markets. All in all, for those who wish to invest without hassle, the classic advice of dollar cost averaging into a broad-market ETF/index fund certainly rings true as demonstrated in the above analyses.
